RNN (Recurrent Neural Network)

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM

data = np.random.random((1000, 10, 1))
labels = np.random.randint(2, size=(1000, 1))

split_ratio = 0.8
split_idx = int(len(data) * split_ratio)
x_train, x_test = data[:split_idx], data[split_idx:]
y_train, y_test = labels[:split_idx], labels[split_idx:]

model = Sequential([
    LSTM(64, input_shape=(10, 1)),
    Dense(1, activation='sigmoid')
])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=10, batch_size=32, validation_data=(x_test, y_test))
loss, accuracy = model.evaluate(x_test, y_test)
print("RNN Accuracy:", accuracy)
```

CNN (Convolutional Neural Network)

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

data = np.random.random((1000, 10, 10, 1))
labels = np.random.randint(2, size=(1000, 1))

split_ratio = 0.8
split_idx = int(len(data) * split_ratio)
x_train, x_test = data[:split_idx], data[split_idx:]
y_train, y_test = labels[:split_idx], labels[split_idx:]

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(10, 10, 1)),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(1, activation='sigmoid')
])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=10, batch_size=32, validation_data=(x_test, y_test))
loss, accuracy = model.evaluate(x_test, y_test)
print("CNN Accuracy:", accuracy)
```

ANN (Artificial Neural Network)

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

data = np.random.random((1000, 10))
labels = np.random.randint(2, size=(1000, 1))

split_ratio = 0.8
split_idx = int(len(data) * split_ratio)
x_train, x_test = data[:split_idx], data[split_idx:]
y_train, y_test = labels[:split_idx], labels[split_idx:]

model = Sequential([
    Dense(64, activation='relu', input_shape=(10,)),
    Dense(1, activation='sigmoid')
])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=10, batch_size=32, validation_data=(x_test, y_test))

loss, accuracy = model.evaluate(x_test, y_test)
print("ANN Accuracy:", accuracy)
```
